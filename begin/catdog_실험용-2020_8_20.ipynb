#Let's develop Keras Model

```
from keras.preprocessing.image import ImageDataGenerator

# ImageDataGenerator는 keras에서 제공하는 이미지데이터 학습을 위한 라이브러리 
# ImageDataGenerator() 함수 사용해서 훈련 데이터를 증강시켜서 예측 결과를 높일 수 있다.
```

```
generator = ImageDataGenerator()

# ImageDataGenerator() : 주어진 데이터 이용해 변형된 이미지 만드는 함수이다.

train_gen = generator.flow_from_directory('dogs-vs-cats/train/train', target_size=(64,64),
                            batch_size = 32,
                            class_mode = 'binary')

# flow_from_directory() : 폴더에 저장된 데이터 불러오는 함수
# dogs-vs-cats/train/train 경로에서 이미지를 읽어와서
# target_size=(64,64)를 통해 사이즈를 resize해주고
# batch_size = 32 한번에 32개씩 묶어주고
# class_mode = 'binary'은 2개의 클래스가 있을 때, 한개는 0, 나머지는 1로 해준다.
# 이를 사용하기 위해서는 기본적으로 경로 아래에 클래스 별로 폴더가 나누어져 있어야 한다. 
# 그래서 train 안에 있는 dog와 cat를 나눠 넣어야 한다. 
```

#Build Model

```
from keras.layers import*
from keras.models import Model

def build_model():
  x =  Input(shape=(64,64,3))
  out = x

  out = Flatten()(out)
  out = Dense(512, activation='relu')(out)
  out = Dense(256,activation='relu')(out)
  out = Dense(64, activation='relu')(out)
  out = Dense(1, activation='sigmoid')(out)

  model = Model(x,out)
  return model

# 딥러닝할 때 사용하는 하나하나의 벡터를 텐서라고 보면 된다.
# Dense는 1차원 벡터에 대해서만 쓸 수 있다.
# 모든 코드를 한줄로 펼쳐야 한다. out = Flatten()(out),
# (shape=(64,64,3))를 한줄로 펼치기 때문에 64 * 64 * 3 약 12000짜리의 크기가 나온다. 
# 이를 Dense를 이용햇 512씩 줄이는 것이다. 
# 입력텐서와 마지막 텐서를 묶어서 Model이라고 한다. from keras.models import Model
# 해당 조건의 shape을 가진 입력 텐서를 만든 것이다. x =  Input(shape=(64,64,3))
# x는 텐서이다. 
# Dense는 A(W* X + b)인 MLP Layer 한층, 다른 말로 fully connected layer 한층을 의미한다.
# 인공신경망의 기본 형태는 Input Layer에서 Output Layer로 연결되는 구조이다. 이를 수식으로 표현하면 아래와 같다.
# y=f(Wx+b)  여기서 x는 입력벡터, b는 편향벡터이며 W는 가중치행렬이다
# out = Dense(512, activation='relu')(out), 512는 output의 크기이다. out = x로 했으니 out을 받고
# layer을 통과한 output을 out으로 저장한다. 
# 여기서는 4층짜리 모델을 만들었다. 
# Dense 레이어는 보통 출력층 이전의 은닉층으로도 많이 쓰이고, 영상이 아닌 수치자료 입력 시에는 입력층으로도 많이 쓰입니다.
# 이 때 활성화 함수로 ‘relu’가 주로 사용됩니다. 
# sigmoid는 0부터1 사이로 mapping한다. 

# model = Model(x,out), 입력 텐서가 x이고 나가는 텐서가 out이다.

# https://blog.naver.com/gyungsumin/221671253777 참고 링크
# https://tykimos.github.io/2017/01/27/MLP_Layer_Talk/ dense relu 참고 링크

```

```
model = build_model()
model.summary()
```

#Let's Train Model

```
# 이 모델을 학습
model.compile(optimizer='sgd', loss='binary_crossentropy')
history = model.fit(train_gen, epochs=10, steps_per_epoch = len(train_gen))

# compile() , 모델 구성 후 compile() 메서드를 호출해서 모델 학습 과정을 설정한다.
# 즉, 모델을 빌드하고 실행하기 전에 컴파일 하는 훈련 준비 단계로 생각하면 된다.
# keras.Model.compile()에는 세 개의 파라미터가 중요하다.
# loss : 최적화 과정에서 최소화 될 손실함수를 설정하는 것으로, MSE(평균 제곱 오차)와 binary_crossentropy가 자주 사용된다
# binary_crossentropy, label을 0과 1로 해놓아서 이를 사용한다.
# optimizer : 훈련 과정을 설정하는 것으로, Adam, SGD 등이 있다
# metrics : 훈련을 모니터링하기 위해 사용한다

# epochs=10, 총 10번 훈련
# steps_per_epoch = len(train_gen), epoch 한번에 몇 번의 batch를 할 것인지 설정
# https://keras.io/ko/optimizers/ - optimizer 설명
```

```
model.compile(optimizer='sgd', loss='binary_crossentropy', metrics=['binary_accuracy'])
history = model.fit(train_gen, epochs=10, steps_per_epoch = len(train_gen))

# test set은 label이 없으니깐 정확도를 못 구하고 train은 label이 있으니 정확도를 구할 수 있다. 
# metrics=['binary_accuracy'], loss외에 추가적으로 확인하고 싶은 지표 값에 대한 함수를 넣을 수 있다. 
# 갈수록 성능이 좋아지지만 이는 overfitting이 될 수도 있다는 점을 유념해야 한다. 
```
