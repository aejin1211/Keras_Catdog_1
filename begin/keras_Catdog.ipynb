```
# 이미지를 0과 1로 매핑
generator = ImageDataGenerator(rescale=1/255.)

train_gen = generator.flow_from_directory('dogs-vs-cats/train/train', target_size=(64,64),
                            batch_size = 32,
                            class_mode = 'binary')
```

#Build Model

```
from keras.layers import*
from keras.models import Model

def build_model():
  x =  Input(shape=(64,64,3))
  out = x

  out = Flatten()(out)
  out = Dense(512, activation='relu')(out)
  out = Dense(256,activation='relu')(out)
  out = Dense(64, activation='relu')(out)
  out = Dense(1, activation='sigmoid')(out)

  model = Model(x,out)
  return model
```

```
model = build_model()
model.summary()
```

#Let's Train Model


```
# 이 모델을 학습
model.compile(optimizer='sgd', loss='binary_crossentropy')
history = model.fit(train_gen, epochs=10, steps_per_epoch = len(train_gen))
```

```
losses = history.history['loss']
accs = history.history['binary_accuracy']

plt.plot(losses, label = 'loss')
plt.plot(accs, label = 'acc')
plt.legend()
plt.show()
# 원래는 validatoin으로 확인해봐야하는데 validation이 없어서
# 보다 이상적인 그래프일 수 있다.
```
